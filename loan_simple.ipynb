```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('/content/Loan_approval_data_2025.csv', encoding = 'latin-1')
```

```python
df = df[(df['age'] >= 19) & (df['age'] < 40) & (df['loan_intent']=='Education')]
df = df.drop(columns = ['customer_id','loan_intent'])
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['occupation_status','product_type'], drop_first=False)
```

```python
df.describe()
```

```python
X = df.drop(columns = ['loan_status'])
y = df['loan_status']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 20, test_size = 0.2, stratify = y)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train,y_train)
```

```python
y_pred = rfc.predict(X_test)
```

```python
rfc.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,y_pred)
```

```python
features = pd.DataFrame(rfc.feature_importances_, index = X.columns)
```

```python
features
```

```python
rfc2 = RandomForestClassifier(n_estimators = 1000, criterion = 'entropy', min_samples_split = 10, max_depth = 14, random_state = 20)
```

```python
rfc2.fit(X_train,y_train)
```

```python
rfc2.score(X_test,y_test)
```

```python
y_pred2 = rfc2.predict(X_test)
```

```python
classification_report(y_test,y_pred2)
```
